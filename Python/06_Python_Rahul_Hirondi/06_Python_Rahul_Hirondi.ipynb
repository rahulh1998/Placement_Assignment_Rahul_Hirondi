 ### ● Get all Pokemons whose spawn time is less than 300 seconds.

### Note - spawn time format is "05:32”, so assume “minute: second” format and perform the analysis.



```python
def time_to_seconds(time):
    if pd.isnull(time):
        return -1
    minutes, seconds = map(int, time.split(':'))
    return minutes * 60 + seconds

# Filter 
spawn_time_threshold = 300
df['Spawn Time (Seconds)'] = df['Spawn Time'].apply(time_to_seconds)
valid_spawn_time_pokemons = df[df['Spawn Time (Seconds)'] >= 0]
spawn_time_less_than_300_pokemons = valid_spawn_time_pokemons[valid_spawn_time_pokemons['Spawn Time (Seconds)'] < spawn_time_threshold]

spawn_time_less_than_300_pokemons

```

```python
import matplotlib.pyplot as plt

# Plotting the distribution of spawn times
plt.figure(figsize=(8, 6))
plt.hist(df['Spawn Time (Seconds)'], bins=20, edgecolor='black')
plt.xlabel('Spawn Time (Seconds)')
plt.ylabel('Count')
plt.title('Distribution of Spawn Times')
plt.show()
```
