## Q3. Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg

```python
mass_above_10000kg = df[df['mass'] > 10000]
mass_above_10000kg
```

```python
plt.figure(figsize=(8, 6))
mass_ranges = pd.cut(mass_above_10000kg['mass'], bins=[10000, 20000, 30000, 40000])
mass_ranges.value_counts().sort_index().plot(kind='bar')
plt.xlabel('Mass Range (kg)')
plt.ylabel('Count')
plt.title('Count of Earth Meteorites with Mass Above 10,000kg in Different Mass Ranges')
plt.show()
```
