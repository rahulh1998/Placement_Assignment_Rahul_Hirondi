```python
import pandas as pd
df = pd.read_csv("PYTHON_05.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.rename(columns={'number': 'episode_no'})
```

```python
df.head()
```

## Q1. Get all the overall ratings for each season and using plots compare the ratings for all the seasons, like season 1 ratings, season 2, and so on.



```python
season_ratings = df.groupby('season')['average_rating'].mean()
season_ratings
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
season_ratings.plot(kind='bar')
plt.xlabel('Season')
plt.ylabel('Average Rating')
plt.title('Average Ratings for Each Season')
plt.show()
```

## Q2. Get all the episode names, whose average rating is more than 8 for every season



```python
high_rated_episodes = df[df['average_rating'] > 8][['name','episode_no']]
high_rated_episodes
```

## Q3.Get all the episode names that aired before May 2019



```python
df['airdate'] = pd.to_datetime(df['airdate'])
before_may_2019_episodes = df[df['airdate'] < '2019-05-01'][['name','episode_no']]
before_may_2019_episodes
```

## Q4.Get the episode name from each season with the highest and lowest rating



```python
highest_rated_episodes = df.groupby('season')['average_rating'].idxmax()
lowest_rated_episodes = df.groupby('season')['average_rating'].idxmin()

highest_rated_names = df.loc[highest_rated_episodes, ['season', 'name']]
lowest_rated_names = df.loc[lowest_rated_episodes, ['season', 'name']]

highest_rated_names
lowest_rated_names
```

## Q5.Get the summary for the most popular ( ratings ) episode in every season

```python
most_popular_episodes = df.groupby('season')['average_rating'].idxmax()
most_popular_summary = df.loc[most_popular_episodes, ['season', 'name', 'episode_no', 'summary']]
most_popular_summary
```
